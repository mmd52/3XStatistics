```python
# Adding ID's
df['Id']=range(1,83)
```

```python
### Next to use all columns i planned they must be in numbers else game over for me
df['SmartPhoneUsagePeriod']=df['SmartPhoneUsagePeriod'].replace(['More than 6 years','Between 4 - 6 years',
                                                                 'Between 2 - 4 years','Less than 2 years'],
                                                                [1,2,3,4])

df['UsagePeriod']=df['UsagePeriod'].replace(['Morning','Afternoon','Evening','Night'],
                                                                [1,2,3,4])

df['NumApps']=df['NumApps'].replace(['More than 60','36 - 60','11 - 35','1 - 10'],[4,3,2,1])

df['Call/Text']=df['Call/Text'].replace(['Text','Call'],[0,1])

df['PriceRange']=df['PriceRange'].replace(['More than $700','$301 - $700','0 - $300'],[3,2,1])
```
